````python
#Part C- Q4

#4.A.
To test if there is a significant difference between funds raised by companies that are still operating vs companies that closed down, we can use a two-sample t-test. Here are the steps to perform this analysis:
1.Import the pandas and scipy libraries and read the CSV file into a dataframe:
    
python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv('CompanyX_EU.csv')

2.Filter the dataframe to only include rows where the "Operating" column is either "Operating" or "Closed":
    
python
df_filtered = df[df['Operating'].isin(['Operating', 'Closed'])]

3.Convert the "Funding" feature to a numerical value:
    
python
df_filtered.loc[:,'Funds_in_million'] = df_filtered['Funding'].apply(lambda x: float(x[1:-1])/1000 if x[-1] == 'K' else (float(x[1:-1])*1000 if x[-1] == 'B' else float(x[1:-1])))

4.Split the dataframe into two groups based on the "Operating" column:
    
python
operating = df_filtered[df_filtered['Operating'] == 'Operating']['Funds_in_million']
closed = df_filtered[df_filtered['Operating'] == 'Closed']['Funds_in_million']

5.Perform the two-sample t-test and get the p-value:
    
python
t, p = ttest_ind(operating, closed, equal_var=False)

#Q4 B.
The null hypothesis is that there is no significant difference between funds raised by companies that are still 
operating vs companies that closed down. The alternative hypothesis is that there is a significant difference between 
the two groups.



#Q4 C
To test for significance and draw a conclusion, we can use the p-value obtained from the two-sample t-test. 
If the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a significant difference 
between funds raised by companies that are still operating vs companies that closed down. If the p-value is greater than or 
equal to 0.05, we fail to reject the null hypothesis and conclude that there is no significant difference between the two groups.
````
